```python
import pandas as pd
from textblob import TextBlob
from textblob.np_extractors import ConllExtractor
```

```python
def phrase_identify(text):
    
    extractor = ConllExtractor()
    blob = TextBlob(text,np_extractor=extractor)

    return (blob.noun_phrases)
```
