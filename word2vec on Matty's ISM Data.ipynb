```python
path1 = "/Users/kumarsanjeev/Downloads/prime_incident"
```

```python
path = os.listdir("/Users/kumarsanjeev/Downloads/prime_incident")
```

```python
list_of_sentences = []
for text in path:
    path2 = path1+'/'+text
    try : 
        data = open(path2 , 'rb')
        for i in data.readlines():
            list_of_sentences.append([i.decode('utf-8').strip()])
    except :
        continue
    
```

```python
list_of_sentences
```

```python
len(list_of_sentences)
```

```python
import re
```

```python
main_list = []
```

```python
for sent in list_of_sentences:
    list2 = []
    for words in sent[0].split():
        if re.match('[^A-Za-z]' ,words) :
            continue
        else :
            if words.startswith('http') or words.startswith('www')  or len(words) < 4 or '@' in words or '\\' in words:
                continue
            else: 
                words = re.sub('[.-?),:;|""(...#]',' ',words.lower())
                list2.append(words.strip())
    main_list.append(list2)
        
```

```python
len(main_list)
```

```python
final_doc = main_list+list1
```

```python
model3 = Word2Vec(final_doc, size=300,min_count=1)
```
