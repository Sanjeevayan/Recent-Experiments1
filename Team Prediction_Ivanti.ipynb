```python

# coding: utf-8

# In[35]:


import numpy as np
import pandas as pd
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier,RidgeClassifier,PassiveAggressiveClassifier
import pickle
from sklearn.metrics import precision_recall_fscore_support,accuracy_score, confusion_matrix, classification_report
from sklearn.calibration import CalibratedClassifierCV
from sklearn.model_selection import train_test_split
from nltk import stem
from sklearn.externals import joblib
from datetime import datetime
import matplotlib.pyplot as plt
import itertools
import nltk 
from nltk import word_tokenize


# In[81]:


data_filepath = "/Users/KumarSanjeev/Desktop/Data.csv"
label_freq_threshold = 100 
cols = ['Subject', 'Symptom', "OwnerTeam"]

stemmer = stem.PorterStemmer()

# In[92]:


def remove_and_dedup_punctuation_numbers(text): 
    result = text.replace('[\w\d\.-]+@[\w\d\.-]+',' ').replace(
    '\d+', ' ').replace('_+', ' ').replace('+', ' ').replace(
    '.', ' ').replace('^https?:\/\/.*[\r\n]*', '').replace('[^\w]',' ').lower()
    result = re.sub (r'([^a-zA-Z\s.?!])', '', result)
    result = re.sub(r"([" + re.escape(string.punctuation) + r"])\1+", r"\1", result)
    result = re.sub( '\s+', ' ', result ).strip()
    return result

def is_weekend(creation_datetime):
    return 1 if creation_datetime.weekday()>4 else 0

def which_weekday(creation_datetime):
    return creation_datetime.weekday()

def clean_text(row):
    sub = row['Subject']
    sym = row['Symptom']

    if row['Subject'] is None:
        sub = ''
    if row['Symptom'] is None:
        sym = ''
    train_text = sub+" "+sym
    text_clean = remove_and_dedup_punctuation_numbers(train_text)
    row["Text_Train_Cleaned"] = text_clean
    
    return row

def encode_decision(y):
    encoder= LabelEncoder()
    encoder.fit(y)
    y_encoded=encoder.transform(y)
    return y_encoded,encoder


# In[95]:
    
class StemmedTfidfVectorizer(TfidfVectorizer):
    def build_analyzer(self):
        analyzer = super(TfidfVectorizer, self).build_analyzer()
        return lambda doc: (stemmer.stem(w) for w in analyzer(doc))


class DataPreprocessor(object):
    def __init__(self):
        pass
    
    @classmethod
    def remove_low_frequent_classes(cls,df, target):
        value_counts = df[target].value_counts() 
        to_remove = value_counts[value_counts <= label_freq_threshold].index
        df[target].replace(to_remove, np.nan, inplace=True)
        
        return df
    
    @classmethod
    def clean_df(cls,df):
        truncated_df_null_removed = df[cols].dropna()
        df_cleaned = truncated_df_null_removed;
        df_cleaned = df_cleaned.apply(clean_text,1)[["Text_Train_Cleaned", "OwnerTeam"]]
        return df_cleaned
      
    @classmethod
    def create_train_test_df(cls,df):
        y = df[["OwnerTeam"]].as_matrix()
        train_dataframe, test_dataframe = train_test_split(df,test_size=0.2,stratify=y,random_state=0)
        return train_dataframe,test_dataframe


# In[77]:


class TeamPredictionModel(object):
    def __init__(self,best_model,best_acc,vectorizer,encoder,cm, cr):
        self.best_model = best_model
        self.best_acc = best_acc
        self.vectorizer = vectorizer
        self.encoder = encoder
        self.cm = cm
        self.cr = cr
    
    @classmethod
    def train(cls,train_dataframe,test_dataframe,target_column):
        vectorizer = StemmedTfidfVectorizer( lowercase=True, sublinear_tf=True,ngram_range=(1,3),min_df=2,max_df=0.95,
                              stop_words='english')

        X_train = vectorizer.fit_transform(train_dataframe["Text_Train_Cleaned"])
        print(len(vectorizer.get_feature_names()) )
        print(X_train.shape)
        
        train_label = list ( train_dataframe[target_column] )
        print(type(train_label))
        Y_train,encoder = encode_decision(train_label)

        X_test = vectorizer.transform(test_dataframe["Text_Train_Cleaned"])
        print(X_test.shape)

        test_label = list(test_dataframe[target_column] )
        Y_test = encoder.transform(test_label)



        clf_LR = LogisticRegression(random_state=0,multi_class='ovr')
        clf_RF = RandomForestClassifier(n_estimators=90, random_state=0, n_jobs=-1,max_depth=11,min_samples_leaf=8)
        clf_SVM = LinearSVC(penalty='l2', tol=1e-4,loss='hinge',max_iter=10,multi_class="ovr",class_weight='balanced')
        clf_SGD_l2 = SGDClassifier(loss='hinge',alpha=.0001, max_iter=10,  penalty='l2')
        clf_SGD_elastic = SGDClassifier(alpha=.0001, max_iter=10,n_jobs=-1, penalty="elasticnet")
        clf_ridge = RidgeClassifier(tol=1e-2, solver="sag",random_state=0)
        clf_pass = PassiveAggressiveClassifier(max_iter=50)
        clf_NB = MultinomialNB()
        classifiers = [clf_LR,clf_RF,clf_SVM,clf_SGD_l2,clf_SGD_elastic,clf_ridge,clf_pass,clf_NB]
        
        best_model = None
        best_acc = -100
        for clf in classifiers:
            model,acc, cf, cr = cls.model_stat(clf,X_train,Y_train,X_test,Y_test,encoder)
            if acc > best_acc:
                best_acc = acc
                best_model = model
                best_cf = cf
                best_cr = cr
            print(type(clf).__name__,": ",acc)
        return TeamPredictionModel(best_model,best_acc,vectorizer,encoder, best_cf, best_cr)
    
    @classmethod
    def model_stat(cls,model,X_train,Y_train,X_test,Y_test, encoder):
        clf_isotonic = CalibratedClassifierCV(model, cv=3, method='isotonic')
        clf_isotonic.fit(X_train, Y_train)
        Y_pred = clf_isotonic.predict(X_test)
        return clf_isotonic,accuracy_score(Y_test, Y_pred), \
            confusion_matrix(encoder.inverse_transform(Y_test), encoder.inverse_transform(Y_pred)), \
            classification_report(encoder.inverse_transform(Y_test), encoder.inverse_transform(Y_pred))
    
    
    def save(self,filepath):
        model_dict = {"model":self.best_model,
                      "accuracy":self.best_acc,
                      "vectorizer":self.vectorizer,
                      "encoder":self.encoder}
        with open(filepath,"wb") as fp:
            pickle.dump(model_dict,fp,pickle.HIGHEST_PROTOCOL)
        


# In[94]:


input_df = pd.read_csv(data_filepath)
c_df = DataPreprocessor.remove_low_frequent_classes(input_df, "OwnerTeam")
c_df = DataPreprocessor.clean_df(input_df)
train,test = DataPreprocessor.create_train_test_df(c_df)
team_pred_model = TeamPredictionModel.train(train,test,"OwnerTeam")


# In[86]:


print(train.__len__() )


# In[68]:


team_pred_model.save("./team_model.pkl")


# In[69]:


model_dict = joblib.load("./team_model.pkl")


# In[73]:


def score(model_dict,subject,symptom):
    text = subject + " " + symptom
    text_cleaned = [remove_and_dedup_punctuation_numbers(text)]
    text_vectorized = model_dict["vectorizer"].transform(text_cleaned)
    result = model_dict["model"].predict_proba(text_vectorized)
    label_index = np.argmax(result)
    label = model_dict["encoder"].inverse_transform(label_index)
    confidence = result[0][label_index]
    response = {"Closure":label,"confidence":confidence}
    return response


# In[74]:


subject = "Request for Former Lumension reporting access"
symptom = "ann.haehn@heatsoftware.com www.HEATsoftware.com (AutoClosed)te instance as well as access to the SSRS report cloud site for Lumension reports. Thank you"


# In[75]:
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    plt.figure(figsize=(25,25))
    plt.gca().xaxis.tick_top()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    #plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

classes = list(c_df['OwnerTeam'].unique())
classes.sort()
plot_confusion_matrix(team_pred_model.cm, classes, normalize=False)

score(model_dict,subject,symptom)
print(team_pred_model.cr)

```
